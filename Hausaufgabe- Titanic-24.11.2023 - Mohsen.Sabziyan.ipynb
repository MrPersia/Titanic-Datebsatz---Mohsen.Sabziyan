**Eine stichpunktartige ErlÃ¤uterung jeder Spalte im Titanic-Datensatz:**

- survived: Gibt an, ob der Passagier Ã¼berlebt hat (1) oder nicht (0).
- pclass: Die Ticketklasse als Proxy fÃ¼r den sozioÃ¶konomischen Status (1 = Erste Klasse, 2 = Zweite Klasse, 3 = Dritte Klasse).
- sex: Geschlecht des Passagiers (mÃ¤nnlich, weiblich).
- age: Alter des Passagiers in Jahren.
- sibsp: Anzahl der Geschwister oder Ehepartner des Passagiers an Bord.
- parch: Anzahl der Eltern oder Kinder des Passagiers an Bord.
- fare: Tarif, den der Passagier bezahlt hat.
- embarked: AbkÃ¼rzung des Hafens, an dem der Passagier an Bord gegangen ist (C = Cherbourg, Q = Queenstown, S = Southampton).
- class: Eine Kategorie, die die Ticketklasse darstellt (Erste, Zweite, Dritte).
- who: Kategorisiert den Passagier als Mann, Frau oder Kind.
- adult_male: Boolescher Wert, der angibt, ob der Passagier ein erwachsener Mann ist.
- deck: Das Schiffsdeck, auf dem sich die Kabine des Passagiers befand (kann fehlende Werte enthalten).
- embark_town: Der Name des Hafens, an dem der Passagier an Bord gegangen ist.
- alive: Gibt an, ob der Passagier Ã¼berlebt hat (â€˜yesâ€™) oder nicht (â€˜noâ€™), Ã¤hnlich wie 'survived', aber in Textform.
- alone: Boolescher Wert, der angibt, ob der Passagier alleine gereist ist (keine Geschwister, Ehepartner, Eltern oder Kinder an Bord).

# Berechnung des Prozentsatzes fehlender Werte:


```python

# 1. Legen Sie Variablen fÃ¼r die Gesamtzahl der Zellen und die Gesamtzahl der fehlenden Daten fest
total_missing = missing_values.sum()

# 2. Berechnen des Prozentsatzes der fehlenden Daten
Prozentsatz_missing = (total_missing/total_cells) * 100

print("Total missing values: {}  =  {:.2f} %".format(total_missing, Prozentsatz_missing))
```

```python
number_missing = df_titanic.isnull().sum().sort_values(ascending=False)
pct_column = (df_titanic.isnull().sum() / len(df_titanic) * 100).round(2).astype('str')+' %'
pct_total = (df_titanic.isnull().sum()/df_titanic.isna().sum().sum()*100).round(2).astype('str')+' %'
missing_values = pd.concat([number_missing, pct_column, pct_total], axis=1, keys=['Number_Missing_Values', 'PCT_Missing_in_Column','PCT_of_all_Missing'])
   
print('\nMISSING VALUES IN',df_titanic.shape[0],'ROWS:')    
all_missing = missing_values.loc[missing_values['Number_Missing_Values'] > 0]
all_missing
```

## 10. Ãœberlebens- und Sterberate nach Einschiffungsort

```python
import matplotlib.pyplot as plt
import seaborn as sns

survival_rate_by_city = df_titanic.groupby('embark_town')['survived'].mean().reset_index()
death_rate_by_city = 1 - survival_rate_by_city['survived']

plt.figure(figsize=(10, 6))

bar_width = 0.4
bar_positions = range(len(survival_rate_by_city))

plt.bar(bar_positions, survival_rate_by_city['survived'], width=bar_width, label='Survival Rate', color='blue')
plt.bar([pos + bar_width for pos in bar_positions], death_rate_by_city, width=bar_width, label='Death Rate', color='red')

plt.xlabel('Embarkation Town', fontsize=12)
plt.ylabel('Rate', fontsize=12)
plt.title('Ãœberlebens- und Sterberate nach Einschiffungsort', fontsize=16, fontweight='bold')
plt.xticks([pos + bar_width / 2 for pos in bar_positions], survival_rate_by_city['embark_town'], rotation=45, fontsize=10)
plt.legend()
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

### 11. Die Anzahl der Ãœberlebenden und Nicht-Ãœberlebenden in verschiedenen Altersgruppen

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definieren der Alterskategorien
def categorize_age(age):
    if age <= 18:
        return '0-18'
    elif age <= 35:
        return '18-35'
    elif age <= 55:
        return '35-55'
    else:
        return '55+'

df_titanic['age_group'] = df_titanic['age'].apply(categorize_age)

data = df_titanic[['survived', 'age_group']]

grouped_data = data.groupby(['survived', 'age_group']).size().reset_index(name='count')

plt.figure(figsize=(10, 6))

sns.barplot(x='age_group', y='count', hue='survived', data=grouped_data, palette='Set2')

plt.xlabel('Altersgruppen', fontsize=12)
plt.ylabel('Anzahl', fontsize=12)
plt.title('Anzahl der Ãœberlebenden und Nicht-Ãœberlebenden in Altersgruppen', fontsize=16, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Survived', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 20. Unterschiedlich die Altersverteilung in den verschiedenen Klassen 

```python
fig = sns.FacetGrid(df_titanic,hue='pclass',aspect=4)

fig.map(sns.kdeplot,'age',shade=True)

oldest = df_titanic['age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
```

#### Ganz deutlich ist zu erkennen, dass die Passagiere der ersten Klasse eher der mittleren Altersgruppe angehÃ¶rten und sehr wenige Kinder hatten. In der zweiten und dritten Klasse gab es mehr Kinder und vergleichsweise weniger Ã¤ltere Menschen.

```python
fig = sns.FacetGrid(df_titanic,hue='sex',aspect=4)

fig.map(sns.kdeplot,'age',shade=True)

oldest = df_titanic['age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
```

#### Die Altersverteilung scheint bei mÃ¤nnlichen und weiblichen Passagieren weniger unterschiedlich zu sein, obwohl wir sagen kÃ¶nnen, dass es mehr jÃ¼ngere Frauen als jÃ¼ngere MÃ¤nner gibt.
